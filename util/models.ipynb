```
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras import backend as K
from tensorflow.python.framework.ops import disable_eager_execution
from keras.regularizers import l2
from tensorflow.python.util import deprecation
deprecation._PRINT_DEPRECATION_WARNINGS = False
K.clear_session()
disable_eager_execution() #To execute the gradient
```

```
tiny_net = models.Sequential(name = "tiny_net")
tiny_net.add(layers.Conv2D(16, (3, 3), activation='relu',
                        input_shape=(150, 150, 1), name = "input_conv"))
tiny_net.add(layers.MaxPooling2D((2, 2), name = "input_max_pooling"))
tiny_net.add(layers.Conv2D(32, (3, 3), activation='relu', name = "conv_1"))
tiny_net.add(layers.MaxPooling2D((2, 2), name = "max_pooling_1"))
tiny_net.add(layers.Conv2D(64, (3, 3), activation='relu', name = "conv_2"))
tiny_net.add(layers.MaxPooling2D((2, 2), name = "max_pooling_2"))
tiny_net.add(layers.Conv2D(64, (3, 3), activation='relu', name = "conv_3"))
tiny_net.add(layers.MaxPooling2D((2, 2), name = "max_pooling_3"))
tiny_net.add(layers.Conv2D(64, (3, 3), activation='relu', name = "conv_4"))
tiny_net.add(layers.MaxPooling2D((2, 2), name = "max_pooling_4"))
tiny_net.add(layers.Flatten(name = "flatten"))
tiny_net.add(layers.Dropout(0.5, name = "drop1"))
tiny_net.add(layers.Dense(512, activation='relu', name = "fully_connected"))
tiny_net.add(layers.Dropout(0.2, name = "drop2"))
tiny_net.add(layers.Dense(1, activation='sigmoid', name = "fully_connected_out"))

print('"tiny_net" CNN loaded' )
```

```
# La regolarizzazione fa underfitting (ricorda di scrivere nella doc)
mammo_net = models.Sequential(name = "mammo_net")
mammo_net.add(layers.Conv2D(16, (3, 3), activation='relu',
                        input_shape=(150, 150, 1), name = "input_conv1"))
mammo_net.add(layers.Conv2D(16, (3, 3), activation='relu', name = "input_conv2"))
mammo_net.add(layers.MaxPooling2D((2, 2), name = "input_max_pooling"))
mammo_net.add(layers.Conv2D(32, (3, 3), activation='relu', name = "block1_conv1"))
mammo_net.add(layers.Conv2D(32, (3, 3), activation='relu', name = "block1_conv2"))
mammo_net.add(layers.MaxPooling2D((2, 2), name = "block1_max_pooling"))
mammo_net.add(layers.Conv2D(64, (3, 3), activation='relu', name = "block2_conv1"))
mammo_net.add(layers.Conv2D(64, (3, 3), activation='relu', name = "block2_conv2"))
mammo_net.add(layers.Conv2D(64, (3, 3), activation='relu', name = "block2_conv3"))
mammo_net.add(layers.MaxPooling2D((2, 2), name = "block2_max_pooling"))
mammo_net.add(layers.Conv2D(128, (3, 3), activation='relu', name = "block3_conv1"))
mammo_net.add(layers.Conv2D(128, (3, 3), activation='relu', name = "block3_conv2"))
mammo_net.add(layers.Conv2D(128, (3, 3), activation='relu', name = "block3_conv3"))
mammo_net.add(layers.MaxPooling2D((2, 2), name = "block3_max_pooling"))
mammo_net.add(layers.Flatten(name = "flatten"))
mammo_net.add(layers.Dropout(0.2, name = "drop1"))
mammo_net.add(layers.Dense(4096, activation='relu', name = "fully_connected_1"))
mammo_net.add(layers.Dropout(0.2, name = "drop2"))
mammo_net.add(layers.Dense(4096, activation='relu', name = "fully_connected_2"))
mammo_net.add(layers.Dropout(0.1, name = "drop3"))
mammo_net.add(layers.Dense(1, activation='sigmoid', name = "fully_connected_out"))

print('"mammo_net" CNN loaded' )
```
